```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

```python
n = 250
p = 0.008
m = 10000
mu = 2
variance = 1.984
sigma = math.sqrt(variance)
```

```python
fig, ax = plt.subplots()

x = np.arange(-2, 6, 1)
dist = [stats.binom.pmf(r, n, p) for r in x ] 
ax.bar(x, dist) 

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
norm, = ax.plot(x, stats.norm.pdf(x, mu, sigma))

x = np.arange(-2, 6, 1)
y = stats.poisson.pmf(x, mu=mu)
ax.plot(x, y)

ax.legend()
plt.show()
```
