```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

```python
n = 7072
p = 0.45
mu = 3182.559
sigma = 41.84
```

```python
fig, ax = plt.subplots()

x = np.arange(3000, 3350, 1)
dist = [stats.binom.pmf(r, n, p) for r in x ] 
ax.bar(x, dist, label = "bino") 

norm = ax.plot(x, stats.norm.pdf(x, mu, sigma), label = "norm", color="red")

y = stats.poisson.pmf(x, mu=mu)
ax.plot(x, y, label = "poi", color="orange")

ax.legend()
plt.show()
```
