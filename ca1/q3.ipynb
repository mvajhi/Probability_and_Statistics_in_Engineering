# part 1

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

```python
mu = 80
sigma = 12
```

```python
fig, ax = plt.subplots()
fig, bx = plt.subplots()

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
norm, = ax.plot(x, stats.norm.pdf(x, mu, sigma))
norm, = bx.plot(x, stats.norm.cdf(x, mu, sigma))

ax.legend()
bx.legend()
plt.show()
```

```python
fig, bx = plt.subplots()

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
norm, = bx.plot(x, stats.norm.cdf(x, mu, sigma))
bx.axhline(y=0.9)

bx.legend()
plt.show()
```

```python
fig, bx = plt.subplots()

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
norm, = bx.plot(x, stats.norm.cdf(x, mu, sigma))
bx.axhline(y=0.9)
bx.axvline(x=95.5)

bx.legend()
plt.show()
```
