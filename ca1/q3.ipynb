# part 1

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

```python
mu = 80
sigma = 12
```

```python
fig, ax = plt.subplots()
fig, bx = plt.subplots()

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
norm, = ax.plot(x, stats.norm.pdf(x, mu, sigma))
norm, = bx.plot(x, stats.norm.cdf(x, mu, sigma))

ax.legend()
bx.legend()
plt.show()
```

```python
fig, bx = plt.subplots()

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
norm, = bx.plot(x, stats.norm.cdf(x, mu, sigma))
bx.axhline(y=0.9)

bx.legend()
plt.show()
```

```python
fig, bx = plt.subplots()

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
norm, = bx.plot(x, stats.norm.cdf(x, mu, sigma))
bx.axhline(y=0.9)
bx.axvline(x=95.5)

bx.legend()
plt.show()
```

# part 4

```python
lam = 17
grads = [i / 4 for i in range(81)]
sample_count = 10000
avg_sample_count = 50000
sample_count_per_avg = 500
```

```python
uniform_sample = np.random.choice(grads, sample_count)
uniform_avg_sample = [np.mean(np.random.choice(uniform_sample, sample_count_per_avg)) for i in range(avg_sample_count)]
plt.hist(uniform_avg_sample, bins=100)
plt.show()
```

```python
exp_samples = np.random.exponential(lam, sample_count)
exp_avg_samples = [np.mean(np.random.choice(exp_samples, sample_count_per_avg)) for i in range(avg_sample_count)]
plt.hist(exp_avg_samples, bins=100)
plt.show()
```

```python
poi_samples = np.random.poisson(lam, sample_count)
poi_avg_samples = [np.mean(np.random.choice(poi_samples, sample_count_per_avg)) for i in range(avg_sample_count)]
plt.hist(poi_avg_samples, bins=100)
plt.show()
```
