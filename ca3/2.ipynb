```python
import matplotlib.pyplot as plt
import numpy as np, random
def set_seed(seed):
    np.random.seed(seed)
    random.seed(seed)
set_seed(810109203)
```

```python
def cal_b1(x, y):
    return (np.sum(x*y)-len(x)*np.mean(x)*np.mean(y)) / (np.sum(x*x)-len(x)*np.mean(x)*np.mean(x))
```

```python
def cal_b0(b1, x, y):
    return np.mean(y)-b1*np.mean(x)
```

```python
def cal_r2(b0, b1, x, y):
    Y = b1 * x + b0
    return 1 - ((np.sum((y - Y)**2)) / (np.sum((y - np.mean(y))**2)))
```

```python
def regression(x, y):
    b1 = cal_b1(x, y)
    b0 = cal_b0(b1, x, y)
    r2 = cal_r2(b0, b1, x, y)
    return b1, b0, r2
```

```python
x_main = np.array([-2.3, -1.1, 0.5, 3.2, 4.0, 6.7, 10.3, 11.5])
y_main = np.array([-9.6, -4.9, -4.1, 2.7, 5.9, 10.8, 18.9, 20.5])
```

```python
def regres_and_plot(x_main, y_main, x_other, y_other, title):
    x = np.array(list(x_main) + list(x_other)) 
    y = np.array(list(y_main) + list(y_other))
    b1, b0, r2 = regression(x, y)
    Y = b1 * x + b0
    print('b0: ', b0)
    print('b1: ', b1)
    print('r2: ', r2)
    plt.scatter(x, y, color = "red") 
    plt.plot(x, Y)
    plt.text(-2, 15, 'b0: ' + str(b0)[:5] + '\n\nb1: ' + str(b1)[:5] + '\n\nr2: ' + str(r2)[:5])
    plt.title(title)
```

```python
regres_and_plot(x_main, y_main, np.array([]), np.array([]), "8 main")
```

```python
regres_and_plot(x_main, y_main, np.array([5.8]), np.array([31.3]), "with outlier")
```

```python
regres_and_plot(x_main, y_main, np.array([20.4]), np.array([14.1]), "with high leverage")
```
