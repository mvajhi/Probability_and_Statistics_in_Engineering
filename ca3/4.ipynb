```python
import matplotlib.pyplot as plt
import numpy as np, random
import statsmodels.api as sm
import scipy.stats as stats

def set_seed(seed):
    np.random.seed(seed)
    random.seed(seed)
```

```python
set_seed(810109203)
random_data = np.random.poisson(3, 5000)
_ = plt.hist(random_data, bins=13)
```

```python
def test(n):
    set_seed(810109203)
    random_data = np.random.poisson(3 * n, 5000)
    statistic, p_value = stats.shapiro(random_data)
    stats.probplot(random_data, dist="norm", plot=plt)
    plt.xlabel('quantiles of norm')
    plt.ylabel('quantiles of data')
    plt.title('Q-Q Plot')
    print(f'n: {n}\nP_value: {p_value:.3f}')
    plt.text(-1, n * 3.0, f'n: {n}\nP_value: {p_value:.3f}', ha='right', va='bottom')
    plt.grid(True)
    plt.show()
```

```python
test(5)
test(50)
test(5000)
```
